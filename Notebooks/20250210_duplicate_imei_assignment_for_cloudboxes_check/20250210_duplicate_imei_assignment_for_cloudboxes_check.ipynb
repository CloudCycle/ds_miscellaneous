The file loaded in below is an export from the TruckDevices Dynamo table. Prior to export the table was filtered only for active trucks, i.e.:
- device_imei exists
- valid_to does not exist

We only filter for active trucks (and records) as we would like the most up-to-date information on imei to CB (based on the serial_no) assignments to be able to check for duplicates
